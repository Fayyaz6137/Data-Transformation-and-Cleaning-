```python
import os
from logging import exception
import pandas as pd
```

### 1. Create a single dataframe with the concatenation of all input csv files, adding a column called country

```python
dfs = []
for item in os.listdir('Data Files'):
    if item.endswith('.csv'):
        try:
            df = pd.read_csv(f'Data Files/{item}', encoding="ISO-8859-1")
            df['country']=item[0:2]
            dfs.append(df)
        except Exception as e:
            print(e)
            print(f'Error : {item}')
dataset = pd.concat(dfs, ignore_index=True)
```

### 2. Extract all videos that have no tag.

```python
dataset.loc[dataset['tags'] == '[none]']
```

### 3. For each channel, determine the total number of views

```python
dataset.groupby('channel_title')['views'].sum().reset_index()
```

### 4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded, and remove those rows from the original dataframe.

```python
excluded = dataset.loc[(dataset['comments_disabled'] == True) & (dataset['ratings_disabled'] == True) | (dataset['video_error_or_removed'] == True)]
dataset = dataset.loc[~(
        (dataset['comments_disabled'] == True) &
        (dataset['ratings_disabled'] == True) |
        (dataset['video_error_or_removed'] == True))
]
```

### 5. Add a like_ratio column storing the ratio between the number of likes and of dislikes

```python
dataset['like_ratio'] = dataset['likes'] / dataset['dislikes']
```

### 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
dataset['publish_time'] = pd.to_datetime(dataset['publish_time'])
dataset['publish_10min'] = dataset['publish_time'].dt.floor('10T')
```

### 7. For each interval, determine the number of videos, average number of likes and of dislikes.

```python
dataset.groupby('publish_10min').agg(
    num_videos=('video_id', 'count'),
    avg_likes=('likes', 'mean'),
    avg_dislikes=('dislikes', 'mean')
).reset_index()
```

### 8. For each tag, determine the number of videos
Notice that tags contains a string with several tags.

```python
dataset['tag_list'] = dataset['tags'].str.replace('"', '', regex=True).str.split('|')
tags_exploded = dataset.explode('tag_list')
tag_counts = tags_exploded.groupby('tag_list')['video_id'].count().reset_index()
tag_counts
```

### 9. Find the tags with the largest number of videos

```python
tag_counts[tag_counts['video_id'] == tag_counts['video_id'].max()]
```

### 10. For each (tag, country) pair, compute average ratio likes/dislikes

```python
tags_exploded.groupby(['tag_list', 'country']).agg(
    avg_like_ratio=('like_ratio', 'mean')
).reset_index()
```

### 11. For each (trending_date, country) pair, the video with the largest number of views

```python
new_data = dataset.loc[dataset.groupby(['trending_date', 'country'])['views'].idxmax()].reset_index(drop=True)
new_data[['trending_date', 'country', 'video_id', 'title', 'views']]
```

### 12. Divide trending_date into three columns: year, month, day

```python
dataset['trending_date'] = pd.to_datetime(dataset['trending_date'], format='%y.%d.%m')
dataset['trending_date_year'] = dataset['trending_date'].dt.year
dataset['trending_date_month'] = dataset['trending_date'].dt.month
dataset['trending_date_day'] = dataset['trending_date'].dt.day
```

### 13. For each (month, country) pair, the video with the largest number of views

```python
new_data = dataset.loc[dataset.groupby(['trending_date_month', 'country'])['views'].idxmax()].reset_index(drop=True)
new_data[['trending_date_month', 'country', 'video_id', 'title', 'views']]
```

### 14. Read all json files with the video categories

```python
dfs = []
for item in os.listdir('Data Files'):
    if item.endswith('.json'):
        try:
            df = pd.read_json(f'Data Files/{item}', encoding='ISO-8859-1')
            df['country']=item[0:2]
            dfs.append(df)
        except Exception as e:
            print(e)
            print(f'Error : {item}')
dataset_json = pd.concat(dfs, ignore_index=True)
```

### 15. For each country, determine how many videos have a category that is not assignable.

```python
def check_assignable(row):
    if row['items']['snippet']['assignable']:
        return 1
    else:
        return 0

dataset_json['assignable_flag'] = dataset_json.apply(check_assignable, axis=1)
dataset_json.groupby('country')['assignable_flag'].sum().reset_index(name='num_of_assignable_videos')
```
